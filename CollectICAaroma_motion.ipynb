```python
import os

baseDir="/nafs/sgreen/fmri/Prisma/subjects"
outputDir="/nafs/sgreen/fmri/Prisma/RestingStateAnalysis"
```

```python
subList=[]
allFileList=os.listdir(baseDir)        
for name in allFileList:
    if ((name[:3]=='ASD') or (name[:2]=='TD') or (name[:3]=='ANX')) and (not name[-3:]=='BET'):
        subList.append(name)
        
# subList=['ANX001', 'ANX002', 'ANX003','ANX004', 'ANX006','ANX007', 'ANX008', 'ANX009', 'ANX010','ANX011']        
print('The sublist is now %s'%subList)
```

## ICA Aroma Numbers

```python
icaDir="restingHCP/ICA_AROMA"   
fileToCreate_ICAaroma="ICAaromaComponentInfo.txt"

fileName1='classification_overview.txt' #File where each row is one component and it's classification as motion/noise or not.
fileName2='classified_motion_ICs.txt' #File with one row listing all the components which were classified as motion/noise
subPotentialMissing=[]
sumComponents=[['SubjectID TotalNumberComponents NumberComponentsKept NumberComponentsRemoved\n']]
numCol=4 #Number of columns in the output text file
```

```python
for subject in subList:
    file1Contents=[]
    file2Contents=[]
        
        
    #Lists which will be filled with the final component data for each subject
    totNumComponents=[]
    numComponentsRemoved=[]
    numComponentsKept=[]
    
      
    try:
        print('The current subject is %s'%subject)
        os.chdir(baseDir)
        os.chdir(subject)
        os.chdir(icaDir)

        #Open the files that will later be saved in one form or another to the .txt files
        fileOpened1 = open(fileName1)
        for item in fileOpened1:
            file1Contents.append(item)
        fileOpened1.close()
            
            
        fileOpened2 = open(fileName2)
        for item in fileOpened2:
            file2Contents=item
        fileOpened2.close()


        totNumComponents=len(file1Contents)-1
        numComponentsRemoved=len(file2Contents.split(','))
        numComponentsKept=totNumComponents-numComponentsRemoved


        if totNumComponents<=0:
            print('Error has occured during subject %s. They appear to have 0 ICA AROMA components. Investigate manually to determine the cause of the error.' %subject)
            subPotentialMissing.append(subject)
            sumComponents.append([subject]+['NaN','NaN','NaN','\n'])                
        else:                              
            sumComponents.append([subject]+[str(totNumComponents)]+[str(numComponentsKept)]+[str(numComponentsRemoved)]+['\n'])

    except:
        print('Error has occured during subject %s. Subject ICA AROMA data may not exist, or another error may have occurred.' %subject)
        subPotentialMissing.append(subject)
        sumComponents.append([subject]+['NaN','NaN','NaN','\n'])
    
```

### Write to text file

```python
n1=1
while n1<len(sumComponents):
    n2=0
    while n2<numCol: #5 is the number of columns there are per subject        
       sumComponents[n1][n2]=str(sumComponents[n1][n2])
       n2+=1
    sumComponents[n1]=' '.join(sumComponents[n1])
    n1+=1
sumComponents[0]=' '.join(sumComponents[0])


#Go to directoy where you want to save results to text file
os.chdir(outputDir)

varFileNew = open(fileToCreate_ICAaroma, 'w')
varFileNew.writelines(sumComponents) #writelines requires a string
varFileNew.close()
```

## Motion

```python

motionDir="restingHCP/1_preprocess/mc" 
fileToCreate_motion="SummaryMotionList.txt"

fileName1='prefiltered_func_data_mcf_abs.rms' #File with one absolute motion value per volume
fileName2='prefiltered_func_data_mcf_abs_mean.rms' #File with single absolute motion value (mean over all volumes)
fileName3='prefiltered_func_data_mcf_rel.rms' #File with one relative motion value per volume
fileName4='prefiltered_func_data_mcf_rel_mean.rms' #File with single relative motion value (mean over all volumes)
subPotentialMissing=[]
allMotionData=[['SubjectID MaxAbsMotion MeanAbsMotion MaxRelMotion MeanRelMotion\n']]
```

```python
for subject in subList:
    #Lists which will be filled with the motion data for each subject
    absMotionList=[]
    absMotionMean=[]
    relMotionList=[]
    relMotionMean=[]


    try:
        print('The current subject is %s'%subject)
        os.chdir(baseDir)
        os.chdir(subject)
        os.chdir(motionDir)

        #Open the files that will later be saved in one form or another to the .txt files
        fileOpened1 = open(fileName1)
        for item in fileOpened1:
            absMotionList.append(float(item))
        fileOpened1.close()

        fileOpened2 = open(fileName2)
        for item in fileOpened2:
            absMotionMean.append(float(item))
        fileOpened2.close()

        fileOpened3 = open(fileName3)
        for item in fileOpened3:
            relMotionList.append(float(item))
        fileOpened3.close() 

        fileOpened4 = open(fileName4)
        for item in fileOpened4:
            relMotionMean.append(float(item))
        fileOpened4.close()         

        absMotionMax=[max(absMotionList)]
        relMotionMax=[max(relMotionList)]

        allMotionData.append([subject]+absMotionMax+absMotionMean+relMotionMax+relMotionMean+['\n'])

    except:
        print('Error has occured during subject %s. Subject motion data from preprocessing may not exist, or another error may have occurred.' % subject)
        subPotentialMissing.append(subject)
        allMotionData.append([subject]+['NaN','NaN','NaN','NaN','\n'])
```

### Write to text file

```python
n1=1
while n1<len(allMotionData):
    n2=0
    while n2<5: #5 is the number of columns there are per subject        
        allMotionData[n1][n2]=str(allMotionData[n1][n2])
        n2+=1
    allMotionData[n1]=' '.join(allMotionData[n1])
    n1+=1
allMotionData[0]=' '.join(allMotionData[0])

os.chdir(outputDir)

varFileNew = open(fileToCreate_motion, 'w')
varFileNew.writelines(allMotionData) #writelines requires a string
varFileNew.close()
```
